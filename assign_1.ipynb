```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

```python
df = pd.read_csv('uber.csv')
```

```python
df
```

```python
print(df.columns)
```

```python
df.info()
```

```python
# 1 pre-process the data set

df.drop(['Unnamed: 0', 'key'],axis=1,inplace=True)
```

```python
print(df.columns)
```

```python
df.fillna(method='ffill',inplace=True)
```

```python
df.drop(df[df['fare_amount'].values <=0].index,inplace=True)
```

```python
df.drop(df[df['passenger_count'] > 10].index,inplace=True)
```

```python
df['pickup_datetime']=pd.to_datetime(df['pickup_datetime'])
```

```python
df['day_of_week']=df['pickup_datetime'].dt.dayofweek
```

```python
missing_values=df.isnull().sum()
```

```python
print("Missing value in the dataset :")
print(missing_values)
```

```python
# 2 Identify outliers
sns.boxplot(x=df['fare_amount'])
plt.show()
```

```python
df.head()
```

```python
# Calculate the IQR for  the 'fare amount' column
q1=df['fare_amount'].quantile(0.25);
q3=df['fare_amount'].quantile(0.75);
IQR=q3-q1

# define threshold
lower_bound=q1-1.5*IQR
upper_bound=q3+1.5*IQR

#remove outliers
data_no_outliers=df[(df['fare_amount']>= lower_bound) & (df['fare_amount'] <=upper_bound)]

sns.boxplot(x=data_no_outliers['fare_amount'])
plt.show()
```

```python
X=df.iloc[:, 2:]
Y=df.iloc[:,0]
print(X.shape,Y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y, test_size=0.15,random_state=42)
```

```python
linear_regression=LinearRegression().fit(x_train,y_train)
```

```python
linear_regression.score(x_test,y_test)
```

```python
y_pred=linear_regression.predict(x_test)
```

```python
result=pd.DataFrame()
result['Actual'],result['Prdeicted']=y_test,y_pred
result.head()
```

```python
print("root mean squared error :",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print("R squared :",metrics.r2_score(y_test,y_pred))
```

```python
#Randfom forestRegressor
random_forest=RandomForestRegressor(n_estimators=10, random_state=10)
```

```python
random_forest.fit(x_train,y_train)
```

```python
random_forest.score(x_test,y_test)
```

```python
y_pred=random_forest.predict(x_test)

result=pd.DataFrame()
result['Actual'],result['predicted']=y_test,y_pred
result.head()
```
